```python
path = 'results'
```

```python
# Python program to get average of a list
def Average(lst):
    return sum(lst) / len(lst)
```

```python
#code to try and read files and then convert to .csv file

import glob
import subprocess


files_sca = glob.glob(path + '/**/*.sca', recursive=True)

for file in files_sca:
    name_run = file
    csv_name = name_run + ".csv"
    cmd = "scavetool x " +  name_run + " -o " + csv_name
    print(cmd)
    
    p = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE,
                             universal_newlines=True,
                             shell=True)

    output, errors = p.communicate()
    exit_code = p.returncode
    print(output,errors,exit_code)
print("Done")
```

```python
import re

def add_values(lstvalues, value):
    if pd.isna(value):
        lstvalues.append(0)
    else:
        lstvalues.append(value)
    
    #print("add_values:",lstvalues)
    



files_sca_csv = glob.glob(path + '/**/*.sca.csv', recursive=True)

df_rows = []
df_cols = []
for file in files_sca_csv:
    #print(file)
    f_split = file.split('/')
    #print(f_split)
    #col_name=f_split[2] + "_" + f_split[3] + "_" + f_split[4]
    col_name=f_split[1] + "_" + f_split[2]
    #print(col_name)
    if col_name not in df_cols:
        df_cols.append(col_name)
        #print(df_cols)
    
    row_name = f_split[-2]
    num = re.findall('\d+',row_name)
    row_name = num[0]
    if row_name not in df_rows:
        df_rows.append(row_name)
        #print(df_rows)
    
    
print("columns", df_cols)
df_rows.sort(key=int)
print("rows",df_rows )
#df_rows.insert(3, df_rows.pop(1))

df2 = pd.DataFrame(index=df_rows, columns=df_cols)
print(df2)

prev_col_name = ""
prev_row_name = ""
lstvalues = []

for file in files_sca_csv:
    f_split = file.split('/')
    col_name=f_split[1] + "_" + f_split[2]
    #col_name=f_split[2] + "_" + f_split[3] + "_" + f_split[4]
    row_name = f_split[-2]
    num = re.findall('\d+',row_name)
    row_name = num[0]

    df = pd.read_csv(file)
    #earstat = df[(df.type=='statistic') & (df.name=='EAR:stats')]
    earstat = df[(df.type=='statistic') & (df.name=='objAge:stats')]
    #earstat = df[(df.type=='statistic') & (df.name=='EteDelay:stats')]
    #earstat = df[(df.type=='statistic') & (df.name=='periodicity:stats')]
    #earstat = df[(df.type=='statistic') & (df.name=='grantWastageMode4:stats')]
    
    values = earstat['mean']
    #print(col_name, row_name, values.mean())
    
    if not prev_col_name and not prev_row_name:
        #print("empty col and row name")
        prev_col_name = col_name
        prev_row_name = row_name
        add_values(lstvalues, values.mean())
    elif prev_col_name == col_name and prev_row_name == row_name:
        #print("-----------------------------------same row and col append data")
        add_values(lstvalues, values.mean())
    else:
        #print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx resetting row and col")
        prev_col_name = col_name
        prev_row_name = row_name
        lstvalues.clear()
        add_values(lstvalues, values.mean())
 
    #print(lstvalues)
    df2.at[row_name,col_name] = Average(lstvalues)

        
print(df2)

"""
df2.plot.line()
plt.ylabel('Grant Wastage')
plt.xlabel('Vehicle Penetration Rate')
plt.show()
#"""    
print("Done")
```
